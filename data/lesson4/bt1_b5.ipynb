```python
import numpy as np 
x = np.random.randint(1,100,100)
print(x)
print(type(x))
```

```python
import matplotlib.pyplot as plt
plt.plot(x)
plt.title('do thi goc x')
plt.show()
```

```python
def find_max(ds):
    max_value = ds[0]
    for num in ds:
     if num > max_value:
        max_value = num
    return max_value
list_data = x 
max_value_data = find_max(list_data)
print("gia tri lon nhat cua day so la : ", max_value_data)
```

```python
a = max(list_data)
print(a)
```

```python
def find_min(ds):
    min_value = ds[0]
    for num in ds:
        if num < min_value:
            min_value = num
    return min_value
min_value_data = find_min(list_data)
print("gia tri nho nhat la: ", min_value_data)
```

```python
def normalization(list_data):
    normal_list = []
    for i in list_data:
        a = (i - min_value_data)/(max_value_data - min_value_data)
        normal_list.append(a)
    return normal_list
a = normalization(list_data)
print(a)
```

```python
print(type(list_data))
```

```python
#tinh normalization

list_data_nor= (list_data - min_value_data)/(max_value_data - min_value_data)
print(list_data_nor)
```

```python
#do thi normalization

plt.plot(list_data_nor)
plt.title("x sau khi normalization")
plt.show()
```

```python
#tinh mean c1

def tinh_mean_c1(listx):
    sum =  0
    for i in listx:
        sum += i 
    meanx = sum/len(listx)    
    return meanx

tinh_mean_c1(x)     
```

```python
#tinh mean c2 

def tinh_meanc2(listx):
    sum = np.sum(listx)
    meanx = sum/len(listx)
    return meanx

tinh_meanc2(x)
```

```python
#tinh mena c3

def tinh_meanc3(listx):
    sizex = np.size(listx)
    sum = np.sum(listx)
    meanx = sum/sizex    
    return meanx

tinh_meanc3(x)  
```

```python
#do lech chuan

def standard(data):
    size = np.size(data)
    y = np.sum((data - tinh_meanc3(data))**2)
    z = y / size
    std = np.sqrt(z)

    return std

result = standard(x)
print(result)
```

```python
# tinh standardlization


def Standardization(data):
    Stand_data = (data - tinh_meanc3(data))/standard(data)
    return Stand_data

result1 = Standardization(x)
result1
```

```python
plt.plot(result1)
plt.title('x sau khi standardlization')
plt.show()
```

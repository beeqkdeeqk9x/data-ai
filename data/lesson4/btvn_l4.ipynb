```python
import numpy as np
x = np.random.randint(1,100,100)
print(x)
print(type(x))
```

```python
import matplotlib.pyplot as plt

```

```python
plt.plot(x)
plt.title('data')
plt.show() 
```

```python
def find_max(ds):
    max_value = ds[0] 
    for num in ds:
     if num > max_value:
        max_value = num
    return max_value
list_data = x 
max_value_data = find_max(list_data)
print("Giá trị lớn nhất trong dãy số là:", max_value_data)
```

```python
def find_min(ds):
    min_value = ds[0] 
    for num in ds:
     if num < min_value:
        min_value = num
    return min_value
min_value_data = find_min(list_data)
print("Giá trị nho nhất trong dãy số là:", min_value_data)
```

```python
def normalization(list_data):
    normal_list = []
    for i in list_data:
        a = (i-min_value_data)/(max_value_data - min_value_data)
        normal_list.append(a)
    return normal_list
a = normalization(list_data)
print(a)
```

```python
list_data_nor = (list_data-min_value_data)/(max_value_data - min_value_data)
print(list_data_nor)
```

```python
plt.plot(list_data_nor)
plt.title('Dataa')
plt.show()
```
